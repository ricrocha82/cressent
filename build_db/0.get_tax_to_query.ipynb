```python
import pandas as pd
```

Spreadsheet downloaded from the ICTV website

```python
path = "/fs/project/PAS1117/ricardo/ssDNA_tool/ssDNA_annotator/test_data/tax_virus_ds_ssdna.csv"
```

```python
df = pd.read_csv(path, sep = ",")
df.head
```

```python
df_ssdna = df[df['Genome'].str.contains("ssDNA", na=False)]
```

```python
names_to_query = df_ssdna[["Family"]].drop_duplicates()
```

```python
names_to_query
```

Remove Pleolipoviridae (contains both dsDNA and ssDNA) and add Alphapleolipovirus

```python
# remove Pleolipoviridae
names_to_query = names_to_query[names_to_query['Family' == "Pleolipoviridae"]]
# add Alphapleolipovirus
new_row = {'Family':'Alphapleolipovirus'}
names_to_query.loc[len(names_to_query)] = new_row
```

```python
names_to_query
```

```python
names_to_query.to_csv('/fs/project/PAS1117/ricardo/ssDNA_tool/ssDNA_annotator/test_data/names_to_query_ssdna.csv', index=False, header=False)
```
